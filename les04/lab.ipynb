6. Правда ли, что люди моложе 30 лет выживали чаще, чем люди старше 60 лет? Каковы доли выживших в обеих группах?

```python
younger = titanic_df[titanic_df['Age'] < 30]
older = titanic_df[titanic_df['Age'] > 60]
ysurvived = round(younger['Survived'].mean(), 2)
osurvived = round(older['Survived'].mean(), 2)
print("Доля выживанания людей моложе 30 лет:", ysurvived)
print("Доля выживания людей старше 60 лет:", osurvived)
print("Люди моложе 30 лет выживали чаще!") if ysurvived > osurvived else print("Данное утверждение неверно")
```

    В среднем мужчины 1-го класса старше 40 лет
    В среднем женщины 1-го класса старше 40 лет
    Мужчины всех классов в среднем старше женщин того же класса
    В среднем люди в 1 классе старше, чем во 2-ом, а те старше представителей 3-го класса



```python
mean_age = titanic_df.groupby(['Pclass', 'Sex'])['Age'].mean()
print(mean_age)

men_first_class = titanic_df[(titanic_df['Pclass'] == 1) & (titanic_df['Sex'] == 'male')]
women_first_class = titanic_df[(titanic_df['Pclass'] == 1) & (titanic_df['Sex'] == 'female')]

men_second_class = titanic_df[(titanic_df['Pclass'] == 2) & (titanic_df['Sex'] == 'male')]
women_second_class = titanic_df[(titanic_df['Pclass'] == 2) & (titanic_df['Sex'] == 'female')]

men_third_class = titanic_df[(titanic_df['Pclass'] == 3) & (titanic_df['Sex'] == 'male')]
women_third_class = titanic_df[(titanic_df['Pclass'] == 3) & (titanic_df['Sex'] == 'female')]


men_first_class_age = round(men_first_class['Age'].mean(), 2)
women_first_class_age = round(women_first_class['Age'].mean(), 2)

men_second_class_age = round(men_second_class['Age'].mean(), 2)
women_second_class_age = round(women_second_class['Age'].mean(), 2)

men_third_class_age = round(men_third_class['Age'].mean(), 2)
women_third_class_age = round(women_third_class['Age'].mean(), 2)


print("\nАнализ утверждений:")


print("1. В среднем мужчины 1-го класса старше 40 лет:", "Верно!" if men_first_class_age > 40 else "Неверно!")


print("2. В среднем женщины 1-го класса старше 40 лет:", "Верно!" if women_first_class_age > 40 else "Неверно!")


men_older_women = (
    men_first_class_age > women_first_class_age and
    men_second_class_age > women_second_class_age and
    men_third_class_age > women_third_class_age
)

print("3. Мужчины всех классов в среднем старше женщин того же класса:", "Верно!" if men_older_women else "Неверно!")


first_class_older = (men_first_class_age > men_second_class_age > men_third_class_age and
                     women_first_class_age > women_second_class_age > women_third_class_age)
print("4. В среднем люди в 1 классе старше, чем во 2-ом, а те старше представителей 3-го класса:", "Верно!" if first_class_older else "Неверно!")
```

10. Сравните возраст у спасенных и у погибших пасажиров. Средний возраст погибших выше, верно?

```python
survived = titanic_df[titanic_df['Survived'] == 1]
dead = titanic_df[titanic_df['Survived'] == 0]
survived_age = round(survived['Age'].mean(), 2)
dead_age = round(dead['Age'].mean(), 2)
print("Возраст спасенных пасажиров:", survived_age)
print("Возраст погибших пасажиров:", dead_age)
print("Средний возраст погибших выше!") if dead_age > survived_age else print("Утверждение неверно")
```

```python
Возраст спасенных пасажиров: 28.34
Возраст погибших пасажиров: 30.63
Средний возраст погибших выше!
```
